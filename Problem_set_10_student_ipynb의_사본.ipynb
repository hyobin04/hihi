```python
import wikipedia 
wiki = wikipedia.page('Artificial intelligence' ) 
text = wiki.content
```

```python
# 여기에 top10_count() 함수 정의 (10점)
def top10_count():
  word_dict = {} 
  for line in text.split('\n')[:10] : 
    word_list = line.split(' ')
    for a in word_list:
     if '=' in a or (not a):
      continue
     else:
      clean_word = a.strip('();,.""')    
     if clean_word not in word_dict:
       word_dict[clean_word] = 1
     else:
       word_dict[clean_word] += 1
  return word_dict
top10_count()
```

```python
import matplotlib.pyplot as plt
def top10_count():
    word_dict = {}
    for line in text.split('\n')[:10]:
        word_list = line.split(' ')
        for a in word_list:
            if '=' in a or (not a):
                continue
            else:
                clean_word = a.strip('();,.')
            if clean_word not in word_dict:
                word_dict[clean_word] = 1
            else:
                word_dict[clean_word] += 1
    
    sorted_count_dict = dict(sorted(word_dict.items(), key=lambda item: item[1], reverse=True))
    top10_words = list(sorted_count_dict.keys())[:10]
    top10_counts = list(sorted_count_dict.values())[:10]
    
    plt.pie(top10_counts, labels=top10_words, autopct='%.3f')
    plt.legend(loc=(1.2, 0.5))
    plt.show()
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].bar(top10_words[:10], top10_counts[:10])
    axes[0].set_xlabel('word')
    axes[0].set_ylabel('number of uses')
    axes[0].set_title('top 10 words count')
    axes[1].boxplot(top10_counts, vert=False)
    axes[1].set_xlabel('number of uses')
    axes[1].set_title('total number of words used')
top10_count()
```
